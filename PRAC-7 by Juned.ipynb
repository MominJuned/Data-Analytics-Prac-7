```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Generate Customer Dataset

```python
np.random.seed(60)

data = {
    "Customer_ID": range(1,501),
    "Name": ["Customer_"+str(i) for i in range(1,501)],
    "Age": np.random.randint(18,65,500),
    "Income": np.random.randint(20000,90000,500),
    "Region": np.random.choice(["North","South","East","West"],500)
}

df = pd.DataFrame(data)

df.head()
```

```python
df.shape
```

### Data Privacy (Mask Sensitive Data)

```python
# Sensitive data like names or personal identifiers should be protected

df["Masked_Name"] = df["Name"].apply(lambda x: x[:3] + "***")

df.head()
# Example: Masking names.
```

```python
# This technique protects user identity.
```

### Data Integrity Check

```python
# Ensure dataset has no missing or incorrect values.

df.isnull().sum()
df.duplicated().sum()
```

```python
# Remove duplicates if present.

df = df.drop_duplicates()
```

### Data Security (Access Limitation)

```python
# Example: Only allow non-sensitive columns for analysis.

secure_df = df[["Customer_ID","Age","Income","Region"]]

secure_df.head()

# This removes personally identifiable information (PII).
```

### Data Analysis on Secure Dataset

```python
# Average income.
secure_df["Income"].mean()
```

```python
# Region wise income.
secure_df.groupby("Region")["Income"].mean()
```

## Visualization — Income Distribution

```python
import numpy as np

income = secure_df["Income"].values

plt.figure()

plt.violinplot(income, showmeans=True)

plt.title("Income Distribution (Violin Plot)")
plt.ylabel("Income")

plt.show()
```

```python
# ---Insight---

# Shows spread of income

# Middle area shows most common income range

# Mean line shows average income
```

```python
# Income Distribution Line Density Graph
```

```python
counts, bins = np.histogram(secure_df["Income"], bins=30)

bin_centers = 0.5*(bins[1:] + bins[:-1])

plt.plot(bin_centers, counts)

plt.title("Income Density Trend")
plt.xlabel("Income")
plt.ylabel("Density")

plt.show()
```

```python
# ---Insight---

# Shows income trend smoothly

# Peaks show most frequent income levels
```

```python
# Income vs Age Bubble Chart
```

```python
plt.scatter(
    secure_df["Age"],
    secure_df["Income"],
    s=secure_df["Income"]/200,
    alpha=0.5
)

plt.title("Age vs Income Bubble Chart")
plt.xlabel("Age")
plt.ylabel("Income")

plt.show()
```

```python
# ---Insight---

# Bubble size shows income magnitude

# Helps detect high earning age groups
```
